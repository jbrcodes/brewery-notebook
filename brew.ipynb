# Scraping US Craft Brewery Statistics

First, import required modules

```python
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

Fetch the HTML page and parse it with BeautifulSoup

```python
url = 'https://www.brewersassociation.org/statistics/by-state/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
```

Get the states

```python
states = soup.find_all('div', attrs={'id': 'state-stats'})
```

Create a DataFrame

```python
columns = ('State', 'Breweries', 'Rank', 'Barrels', 'Rank')
brew_df = pd.DataFrame(columns=columns)
```

Add a few stats for each state to the DataFrame

```python
for state in states:
    name = state.find('h1').get_text()
    if name == 'DC':
        continue
    
    # Get number of breweries and ranking
    count = state.select('#per-state .total .count')[0].get_text()
    rankText = state.select('#per-state .total .rank')[0].get_text()
    countRank = re.search(r'Ranks (\d+)', rankText).group(1)
    
    # Get number of barrels and ranking
    barrels = state.select('#production .total .count')[0].get_text()
    rankText = state.select('#production .total .rank')[0].get_text()
    barrelsRank = re.search(r'Ranks (\d+)', rankText).group(1)
    
    # Create frame for this state and append to main one
    stats = (name, count, countRank, barrels, barrelsRank)
    temp_df = pd.DataFrame(stats).transpose()
    temp_df.columns = columns
    brew_df = pd.concat([brew_df, temp_df], ignore_index=True)
```
